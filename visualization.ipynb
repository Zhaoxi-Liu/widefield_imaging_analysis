```python
def plot_traces(data, data_rate=None, xlim=None):
    '''
    Plot the traces of the data.
    data: np.array, shape=(nframes, ntrials)
    '''
    figsize = (20, 5)
    fig, ax = plt.subplots(figsize=figsize)
    if data_rate is not None:
        x = np.arange(data.shape[0]) / data_rate
    else:
        x = np.arange(data.shape[0])

    for i in range(data.shape[1]):
        ax.plot(x, data[:, i], label='trial {}'.format(i))

    if data_rate is not None:
        ax.set_xlabel('Time (s)')

    if xlim is not None:
        ax.set_xlim(left=xlim[0], right=xlim[1])
    else:
        ax.set_xlim(left=-x[-1]*0.01, right=x[-1]*1.01)    
    ax.legend()
    plt.show()

plot_traces(mean_frames, data_rate=10)
```
